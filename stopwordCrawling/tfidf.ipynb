```python
from glob import glob
import pandas as pd
import numpy as np

paths = sorted(glob('Bows/video*'), key=lambda x: [len(x), x])
wordList = open('Bows/wordList.txt', 'rt')
words = wordList.readlines()
existWords = []
```

```python
for word in words:
    existWords.append(word.strip())

data = pd.DataFrame(columns=existWords)
for path in paths:
    data.loc[path[:-4]] = 0
data
```

```python
for path in paths:
    with open(path, 'rt') as bowFile:
        wordCounts = bowFile.readlines()
        for wordCount in wordCounts:
            line = wordCount.split()
            data.loc[path[:-4], line[0]] += int(line[1])
data
```

```python
df = data.copy()
```

```python
tf = df.div(df.sum(axis=1), axis=0)

# IDF 계산
# 각 단어가 등장한 문서 수 계산
doc_frequency = (df > 0).sum(axis=0)
num_docs = len(df)
idf = np.log(num_docs / (1 + doc_frequency))

# TF-IDF 계산
tf_idf = tf * idf

# 단어별 TF-IDF 평균 계산
word_mean_tfidf = tf_idf.mean(axis=0)

# 중요도가 낮은 단어의 임계값 설정
threshold = 0.000007  # 임계값 설정

# 중요도가 낮은 단어의 목록 구하기
low_importance_words = word_mean_tfidf[word_mean_tfidf < threshold].index.tolist()
high_importance_words = word_mean_tfidf[word_mean_tfidf >= threshold].index.tolist()
print("len low importance words: ", len(low_importance_words))
print("len high importance words: ", len(high_importance_words))
print("Low Importance Words:", low_importance_words)
print("High Importance Words:", high_importance_words)

```

```python
for i in range(1, len(low_importance_words)):
    print(low_importance_words[i], end=' ')
    if i % 10 == 0:
        print()
```

```python
wordBag = set(low_importance_words)
with open('korean_stopwords.txt', 'rt') as stopWord:
    stopwords = stopWord.readlines()
    for word in stopwords:
        wordBag.add(word.strip())

with open('korean_stopwords.txt', 'wt') as stopWord:
    for word in list(wordBag):
        stopWord.write(word)
        stopWord.write('\n')
```
